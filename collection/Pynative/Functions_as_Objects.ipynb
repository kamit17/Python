```python
def is_triangular(n):
    """ n is an int > 0
    Returns True if n is trangular, i.e. equals a continued summation of natural numbers (1+2+3+...+k), False otherwise"""

    total = 0
    for i in range(n+1):
        #print(i)
        total += i
        if total == n:
            return True
    return False

print(is_triangular(4))  #print false
print(is_triangular(6))  #print true
print(is_triangular(1))   #print true
```

```python
def bisection_root(x):
    epsilon = 0.01
    low = 0
    high = x
    ans = (high + low)/2.0
    while abs(ans**2 - x) >= epsilon:
        if ans**2 < x:
            low = ans
        else:
            high = ans
        ans = (high + low)/2.0
    return ans

#print(bisection_root(123))
```

```python
def count_nums_with_sqrt_close_to(n,epsilon):
    """ n is an int > 2
        epsilon is a positive number < 1
    Retunrs how many integers have a square root within  n
    """
    count = 0
    for i in range(n**3):
        #take the square root of i
        sqrt = bisection_root(i)
        #check if sqrt is close to n 
        if abs(n-sqrt) < epsilon:
            print(i,sqrt)
            #know that sqrt is within epsilon
            count += 1
    return count
            
#print(count_nums_with_sqrt_close_to(10,0.10))
print(count_nums_with_sqrt_close_to(10,1))
```

```python
def apply(criteria,n):
    """ criteria is a function that takes in a number and returns a Boolean
        n is an int
    Returns how many ints from 0 to n (inclusive) match the criteria 
    (i.e. return True when criteria is applied to them)
    """ 
    count = 0
    for i in range(n+1):
        #criteria(i)  #is a boolean
        if criteria(i):
            count += 1
    return count
    


def is_even(x):
    return x%2==0

how_many = apply(is_even,10)
print(how_many)
```
