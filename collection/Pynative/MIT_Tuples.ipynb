```python
#return more than one value from a function

def quotient_and_remainder(x,y):
    q = x // y
    r = x % y
    return(q,r)

both = quotient_and_remainder(10,3)
print(both)
(quot,rem) = quotient_and_remainder(5,2)
print(quot)
print(rem)
```

```python
def char_counts(s):
    """s is string of lowercase chars
    Return a tuple where the first element is the
    number of vowels in s and the second element is
    the number of consonants in s """

    vowels = "aeiou"
    (c,v) = (0,0)
    for char in s:
        #char is 'a', then 'b','c'
        if char in vowels:
            v += 1
        else:
            c += 1
    return (v,c)

print(char_counts("abcd"))   
    
```
