```python
def recurring(value):
    if value >= 10:
        print("I'm stopping at iteration {}".format(value))
        return value
    else:
        print("I'm at iteration {}".format(value))
        value += 1
        print("Changed value to {}".format(value))
        print("*********recursive function activate*********")
        print("                                             ")
        recurring(value)
x = 0
recurring(x)
```

```python
def factorial(n):
    """This function returns the factorial of a given number."""
    if n == 0:   #base Case
        return 1
    else:            #recursive case
        return n * factorial(n-1)
        
#print(factorial.__doc__)
print(factorial(5))
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(5))
```

```python
def addition(num):
    if num > 0:
        return num + addition(num - 1)
    else:
        #base case
        return 0
print(addition(10))
```

```python
def is_even(list1):
    even_num =  []
    for n in list1:
        if n % 2 == 0:
            even_num.append(n)
    #return a list
    return even_num

#pass the list to the functions
even_num = is_even([2,3,42,51,62,70,5,9])
print("Even numbers are: ",even_num)
```

```python
#global keyword
#Global variable
#global is the keyword used to access the actual global variable from outside the function. we use the global keyword for two purposes:

#    To declare a global variable inside the function.
#    Declaring a variable as global, which makes it available to function to perform the modification.

x = 5

def function1():
    print("Value in 1st function :",x)
def function2():
    #Modify global variable using global keyword
    global x
    x = 555
    print("Value in second function:",x)

def function3():
    print("Value in third function:",x)
    
function1()
function2()
function3()
```

```python
#nonlocal variable
def outer_func():
    x = 777

    def inner_func():
        #local variable now acs as a global variable
        nonlocal x
        x = 700
        print("value of x inside inner function is :",x)

    inner_func()
    print("value of x inside outer function is: ",x)

outer_func()
        
```

```python
#variable length arguments -- pass multiple numbers of arguments to the function

def addition(*numbers):
    total = 0
    for num in numbers:
        total = total + num
    print("Sum is:",total)

#0 arguments
addition()

#5 arguments
addition(1,4,3,5,6)
```
