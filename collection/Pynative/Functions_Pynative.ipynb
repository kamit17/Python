```python
def outer_func(a,b):
    def inner_func(a,b):
        return a + b

    #call inner function and add 5 to the result
    add = inner_func(a,b)
    return add + 5

print(outer_func(5,10))
```

```python
def recurring(value):
    if value >= 10:
        print("I'm stopping at iteration {}".format(value))
        return value
    else:
        print("I'm at iteration {}".format(value))
        value += 1
        print("Changed value to {}".format(value))
        print("*********recursive function activate*********")
        print("                                             ")
        recurring(value)
x = 0
recurring(x)
```

```python
def factorial(n):
    if n == 0:   #base Case
        return 1
    else:            #recursive case
        return n * factorial(n-1)
print(factorial(5))
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(5))
```

```python
def addition(num):
    if num > 0:
        return num + addition(num - 1)
    else:
        #base case
        return 0
print(addition(10))
```
