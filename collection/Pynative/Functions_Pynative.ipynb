```python
def func1(*args):
    print("Printing values:")
    for i in args:
        print(i)
        
func1(20,40,60)
func1(80,100)
```

```python
def calculation(a,b):
    add = a + b
    sub = a - b
    return add,sub

print(calculation(40,10))
```

```python
def outer_func(a,b):
    def inner_func(a,b):
        return a + b

    #call inner function and add 5 to the result
    add = inner_func(a,b)
    return add + 5

print(outer_func(5,10))
```

```python
def recurring(value):
    if value >= 10:
        print("I'm stopping at iteration {}".format(value))
        return value
    else:
        print("I'm at iteration {}".format(value))
        value += 1
        print("Changed value to {}".format(value))
        print("*********recursive function activate*********")
        print("                                             ")
        recurring(value)
x = 0
recurring(x)
```
