```python
def apply(criteria,n):
    """
    * criteria function that takes in a number and retunrs a bool
    * n: an int
    Returns how many ints from 0 to n (inclusive) match the criteria(i.e return True when run with criteria)
    """

    count = 0
    for i in range(n+1):
        if criteria(i):
            count += 1
        return count

def is_even(x):
    return x % 2 == 0

print(apply(is_even,10))
```

```python
def do_twice(n,fn):
    return fn(fn(n))

print(do_twice(3,lambda x: x**2))
```
