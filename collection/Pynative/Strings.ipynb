```python
#Write a program to create a new string made of an input string’s first, middle, and last character.

input_string = "james"

first_char = input_string[0]
print(f"first_char is: {first_char}")

middle_index = len(input_string) // 2
middle_char = input_string[middle_index]
print(f"middle char is: {middle_char}")

last_char = input_string[-1]
print(f"last_char is: {last_char}")
print(f"string made from first,middle and last charcters is : {first_char+middle_char+last_char}")

```

```python
#Create a string made of the middle three characters
word = "JhonDipPeta"
middle_index = len(word)//2
middle_char = word[middle_index]
print(f"middle char is: {middle_char}")

#Use string slicing to get the middle three characters starting from the middle index to the next two character
char_before = word[middle_index -1]
print(f"char before middle char is: {char_before}")
char_after = word[middle_index +1]
print(f"char after middle char is: {char_after}")
final_word = char_before + middle_char + char_after
print(f"final_word is : {final_word}")
```

```python
# Append new string in the middle of a given string
str1 = "hello"
str2 = "mello"

#print(len(str1))
mi =int(len(str1)//2)

#get the characters from start till middle index
x = str1[:mi]
#print(x)

#add the characters from start till middle of str1 to str2
x = x + str2

#add the characters from the end of middle index to the last character in str1
x = x + str1[mi:]

print(x)
```

```python
#Create a new string made of the first, middle, and last characters of each input string

#Given two strings, s1 and s2, write a program to return a new string made of s1 and s2’s first, middle, and last characters.
s2 = "Australia"
s1 = "Japan"
# First characters
s1_first = s1[0]
s2_first = s2[0]

# Middle characters
s1_middle = s1[len(s1)//2]
s2_middle = s2[len(s2)//2]

# Last characters
s1_last = s1[-1]
s2_last = s2[-1]

# Combine
new_string = s1_first + s1_middle + s1_last + s2_first + s2_middle + s2_last
print(new_string)
```
