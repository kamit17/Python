```python
#singel item tuple
single_tuple = ("Hello")
print(type(single_tuple))  #type str, not tuple since missing , after Hello
print(single_tuple)
single_tuple1 = ("Hello",)
print(type(single_tuple1))  #type tuple post addition of , after hello
print(single_tuple1)
```

```python
#packing variables into a tuple

tuple1 = 1,2,"Hello"
#display tuple
print(tuple1)
print(type(tuple1))

#unpacking tuple into variable
i,j,k = tuple1
#printing the variables
print(i,j,k)
```

```python
#length of a tuple
tuple1 = ('P','Y','T','H','0','N')
#length of a tuple
print(f"length of tuple is:{len(tuple1)}")

for i in range(len(tuple1)):
    print(tuple1[i])

print(f"Element at index -3 is : {tuple1[-3]}")

for i in range(-6,0):
    print(tuple1[i],end= ", ")

numbers= (0,1,2,33,4,55,6,7,8,9,10,23,44)
print()
print(numbers[1:5])
print(numbers[-5:-1])

#find number intuple

#get index of item 5
print(numbers.index(33))

#find within a range
print(numbers.index(55,4,11))

#check if item exists intuple
print(55 in numbers)
print(555 in numbers)
```

```python
#iterating a tuple
sample_tuple = tuple((1,2,3,4,"Hello",[4,6,7,8]))
#iterate a tuple
for item in sample_tuple:
    print(item)
```

```python
#count the occurence of an item in a tuple using count method
tuple1 = (10,20,30,40,60,30,20,30)

#count all occurneces of 30
count = tuple1.count(30)
print(count)
print(min(tuple1))
```

```python
#copying a tuple 
tuple1 = (10,20,30,40,60,30,20,30)
tuple2 = tuple1
print(tuple2)

#concat two tuples
tuple3 = tuple1+ tuple2
print(tuple3)

#sum of two tuples
tuple4 = sum((tuple1,tuple2),())
print(tuple4)

print(min(tuple4))
```

```python
#nested tuples
nested_tuple = ((20, 40, 60), (10, 30, 50), "Python")

#access the first item of the third tuple
print(nested_tuple[2][0])

#iterate a nested tuple
for i in nested_tuple:
    print("tuple",i,"elements")
    for j in i:
        print(j,end=", ")
    print("\n")
```
