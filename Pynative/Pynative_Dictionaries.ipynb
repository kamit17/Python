```python
#creating a dictionary using {}
person = {"name":"Jesse","Country":"USA","telephone": "86753099"}
print(person)

#create a dictionary using dict()

person = dict ({"name":"Jesse","Country":"USA","telephone": "86753099"})
print(person)
#emptydict
emptydict = {}
print(type(emptydict))
```

```python
#Accessing elements of a dictionary
person = {"name":"Jesse","Country":"USA","telephone": "86753099"}

print("--------All the keys------")
#get all the keys
print(person.keys())
print(type(person.keys()))
print()

print("--------All the key value pairs ------")
#get all  key-value painrs
print(person.items())
print(type(person.items()))
print()

print("--------All the values------")
#get all the values
print(person.values())
print(type(person.values()))
print()
print("--------All the key value pairs ------")
#get all  key-value painrs
print(person.items())
print(type(person.items()))
print()

print("--------Acess Values ------")
#acess value using key name in []
print(person['name'])

#get key value using key name in get()
print(person.get('telephone'))
```

```python
#iterating a dictionary
#iterate using a for loop and access the individual keys and their corresponding values

person = {"name":"Jesse","Country":"USA","telephone": "86753099"}

#iterating the dictionary using for-loop
print('key',':','value')
for key in person:
    print(key,':',person[key])

#using items() method
print("-----------Using items() method-------")
print('key',':','value')
for key,value in person.items():
    print(key,':',value)

#count number of keys present in a dictionary
print(len(person))
print()
print("-----Adding items to the dictionary----")

#update dictionary by adding two new keys
person["weight"] = 50
person.update({"height": 6})

person.update([("city", "Texas"), ("company", "Google",)])

print(person)
```

```python
#set default value to a key
person_details = {"name":"Jesse","Country":"USA","telephone": "86753099"}


#set default value if key doesn't exist
person_details.setdefault('state','Texas')

#key doesn't exist and value not mentioned.default none
person_details.setdefault("ZIP")

#key exists and value mentioned, does'nt change value
person_details.setdefault('Country','Canada')

#display dicitonary
for key,value in person_details.items():
    print(key,':',value)
```

```python
#modify value of the dictionary 
#using keyname  -- directl assign value by using its key name.
#using update method -- pass key-value pari to change the value.

person = {"name": "Jesse","Country":"USA"}

#updating the country name
person["Country"] = "Canada"
print(person['Country'])

#using update method
person.update({"Country":"USA"})

print(person['Country'])
```
