```python
numbers = [10,20,30,40,50]
print(f"Third element is {numbers[2]}")
print(f"Length of the list is {len(numbers)}")
is_empty = len(numbers) == 0
print(f"Is the list empty? {is_empty}")
```

```python
numbers = [100,50,400,500]
#change second element of list to 200
numbers[1] = 200
print(numbers)
#Add 600 to the end of the new list
numbers.append(600)
print(numbers)
numbers.insert(2,300)
print(numbers)
numbers.remove(600)
print(numbers)
numbers.pop(0)
print(numbers)
```

```python
numbers = [10, 20, 30, 40, 50]
sum = 0
for num in numbers:
    sum = sum + num
print(f"Sum {sum}")
avg = sum/(len(numbers))
print(f"Average {avg}")
```

```python
data = [45, 12, 89, 2, 67]
print(f"Maximum: {max(data)}")
print(f"Minimum: {min(data)}")
```

```python
factors = [2, 3, 5, 7]
product  = 1
for i in factors:
    #product = product * i
    product *= i
print(f"Product: {product}")
```

```python
numbers = [10, 21, 4, 45, 66, 93, 11]
odd = []
even = []

for num in numbers:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)
print(f"Even Numbers: {len(even)}")
print(f"Odd Numbers: {len(odd)}")
```
