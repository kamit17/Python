```python
numbers = [10, 21, 4, 45, 66, 93, 11]
odd = []
even = []

for num in numbers:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)
print(f"Even Numbers: {len(even)}")
print(f"Odd Numbers: {len(odd)}")
```

```python
list_data = [23, 65, 19, 90]
idx1, idx2 = 0, 2

print(f"Original: {list_data}")

# Pythonic swap
list_data[idx1], list_data[idx2] = list_data[idx2], list_data[idx1]

print(f"Swapped: {list_data}")
```

```python
my_list = [10, 20, 10, 30, 40, 40, 20, 50]
unique_list = []

for num in my_list:
    if num not in unique_list:
        unique_list.append(num)
print(f"Unique List: {unique_list}")
```
