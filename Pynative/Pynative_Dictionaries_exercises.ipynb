```python
#Count Character Frequencies
#Problem Statement: Write a Python program to count how many times each character appears in a given string, 
#storing the results in a dictionary.

def count_characters(text):
    char_count = {} # dictionary to hold count
    for char in text:
    # if character is already in dictionary , increment its count
        if char in char_count:
            char_count[char] += 1
    #if not , add the charater with a count of 1
        else:
            char_count[char] = 1
    return char_count

text = input("Enter any text: ")
result = count_characters(text)
print(result)
```
