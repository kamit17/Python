```python
L1 = [1, 2, 3, 4, 5]

while True:
    index = int(input("Index: "))
    
    if index == -1:
        break
    
    value = int(input("New value: "))
    L1[index] = value
    print(L1)
```

```python
items = int(input("How many items:"))
List1 = []

for i in  range(items):
    value = int(input(f"Item {i+1}: "))
    List1.append(value)
print(f"New List is {List1}")
```

```python
#Removing items for a list
#pop -- if index of item is known

my_list = [1,2,3,4,5,6]
#my_list.pop(2)
print(my_list)
number = my_list.pop(3)  #returns the removed item
print(number)
print(my_list)
```

```python
##remove -- if the contents of the item are known.
#reove takes the value of the item to be removed as its argument

my_list = [1,2,3,4,5,6]
my_list.remove(2)  #item 2 is removed 
print(my_list)
my_list.remove(5)
print(my_list)

#removes the first occurence of teh value in the list

my_list = [1,2,1,2]
my_list.remove(1)
print(my_list)
```

```python
List1 = []   #Empty list to store the elements
while True:
    print(f"The list is now {List1}")
    #
    operation = input("a(d)d,(r)emove or e(x)it:")
    if operation == "x":
        break
    elif operation == "d":
        number = int(input("number:"))
        List1.append(number)
    elif operation == "r":
        List1.pop()
print("Bye")
```

```python
#check for presence of an item with the in operator
my_list = [1,3,4]
if 1 in my_list:
    print("The list contains item 1")

if 2 in my_list:
    print("The list containers item2")
```

```python
words = []
while True:
    word = input("Word:")
    if word in words:
        print(f"You typed in {len(words)} different words")
        break
    else:
        words.append(word)
 
```

```python
#sorting lists
#sort  modifies the original list -- by just changing the order of the original list in place
my_list = [2,5,1,2,4]
my_list.sort()
print(my_list)

```

```python
#sorted doesn't change the original list.It creates a new ordered copy of the list It returns a sorted list
my_list = [2,5,1,2,4]
print(sorted(my_list))

original = [2,5,1,2,4]
in_order = sorted(original)
print(original)
print(in_order)
```
