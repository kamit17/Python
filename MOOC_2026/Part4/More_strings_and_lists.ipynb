```python
def most_common_character(input_string):
    counter = 0
    #stores the character currently considered the most common
    most_common = ""
    for char in input_string:
        #count how many times the current character appears in the whole string
        repeat = input_string.count(char)
        #compare the current character’s count with the best count seen so far
        if repeat > counter:
            #update the highest count
            counter = repeat
            #replace the current winner with this character
            most_common = char
    return most_common

first_string = "abcdbde"
print(most_common_character(first_string))

second_string = "exemplaryelementary"
print(most_common_character(second_string))
    
```

```python
def no_shouting(upper_list: list):
    new_list = []
#upper_list=[]
    for i in upper_list:
        if not i.isupper():
            new_list.append(i)
    return (new_list)
        
        
my_list = ["ABC", "def", "UPPER", "ANOTHERUPPER", "lower", "another lower", "Capitalized"]  
pruned_list = no_shouting(my_list)
print(pruned_list)
```

```python
def input_from_user(how_many: int):
    print(f"Please type in {how_many} numbers:")
    numbers = []

    for i in range(how_many):
        number = int(input(f"Number {i+1}: "))
        numbers.append(number)

    return numbers

def print_result(numbers: list):
    print("The numbers are: ")
    for number in numbers:
        print(number)

def analyze(numbers: list):
    mean  = sum(numbers) / len(numbers)
    return f"There are altogether {len(numbers)} numbers, the mean is {mean}, the smallest is {min(numbers)} and the greatest is {max(numbers)}"

def main():
    
    # the "main function" using these functions
    inputs = input_from_user(5)
    print_result(inputs)
    analysis_result = analyze(inputs)
    print(analysis_result)

main()
```

```python
def get_inputs():
    results = []

    while True:
        line = input("Exam points and exercises completed: ")

        # Stop reading input when the user enters an empty line
        if line == "":
            break

        # Split the line into two parts and convert both parts to integers
        exam_points, exercises = map(int, line.split())

        # Store one student's data as [exam_points, exercises]
        results.append([exam_points, exercises])

    return results


def exercise_to_points(exercises):
    # 87 exercises becomes 8 points, 55 becomes 5, etc.
    return exercises // 10


def get_grade(exam_points, exercise_points):
    total_points = exam_points + exercise_points

    # If exam points are below 10, the student automatically fails
    if exam_points < 10:
        return 0
    elif total_points <= 14:
        return 0
    elif total_points <= 17:
        return 1
    elif total_points <= 20:
        return 2
    elif total_points <= 23:
        return 3
    elif total_points <= 27:
        return 4
    else:
        return 5


def calculate_statistics(results):
    total_points_list = []
    grades = []

    for exam_points, exercises in results:
        # Convert completed exercises into exercise points
        exercise_points = exercise_to_points(exercises)

        # Total points = exam points + exercise points
        total_points = exam_points + exercise_points

        # Find the final grade for this student
        grade = get_grade(exam_points, exercise_points)

        # Save total points for average calculation
        total_points_list.append(total_points)

        # Save grades for pass percentage and distribution
        grades.append(grade)

    return total_points_list, grades


def points_average(total_points_list):
    return sum(total_points_list) / len(total_points_list)


def pass_percentage(grades):
    passed = 0

    for grade in grades:
        # Grades above 0 count as passed
        if grade > 0:
            passed += 1

    return 100 * passed / len(grades)


def print_grade_distribution(grades):
    print("Grade distribution:")

    # Print grades from 5 down to 0
    for grade in range(5, -1, -1):
        stars = ""

        # Add one star for each student who got this grade
        for g in grades:
            if g == grade:
                stars += "*"

        print(f"  {grade}: {stars}")


def print_statistics(total_points_list, grades):
    print("Statistics:")

    # :.1f means show 1 digit after the decimal point
    print(f"Points average: {points_average(total_points_list):.1f}")
    print(f"Pass percentage: {pass_percentage(grades):.1f}")

    print_grade_distribution(grades)


def main():
    # Read all student results from the user
    results = get_inputs()

    # Build the lists needed for the final statistics
    total_points_list, grades = calculate_statistics(results)

    # Print all final output
    print_statistics(total_points_list, grades)


main()
```
