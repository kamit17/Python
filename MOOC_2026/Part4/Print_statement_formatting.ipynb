```python
def formatted(my_list):
    new_list = []
    for number in my_list:
        formatted_number = f"{number:.2f}"
        new_list.append(formatted_number)
    return new_list

        
my_list = [1.234, 0.3333, 0.11111, 3.446]
new_list = formatted(my_list)
print(new_list)
        
```
