```python
my_list = [3,2,4,5,2]
index = 0
while index < len(my_list):
    print(my_list[index])
    index += 1
```

```python
#with for
my_list = [3,2,4,5,2]
for item in my_list:
    print(item)
```

```python
name = input("Please type in your name: ")

for character in name:
    print(character)
```

```python
name = input("Please type in a string: ")

for character in name:
    print(character)
    print("*")
```

```python
num = int(input("Please type in a positive integer: "))
for i in range(-num, num+1):
    if i != 0:
        print(i)
```

```python
def list_of_stars(numbers: list):
    for num in numbers:
        print("*" * num)
list_of_stars([3,7,1,1,2])
```
