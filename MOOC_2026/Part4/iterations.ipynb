```python
def anagrams(string1,string2):
    return sorted(string1) == sorted(string2)


print(anagrams("tame", "meta")) # True
print(anagrams("tame", "mate")) # True
print(anagrams("tame", "team")) # True
print(anagrams("tabby", "batty")) # False
print(anagrams("python", "java")) # False
```

```python
def palindrome(input_string):
    return input_string == input_string[::-1]

while True:
    string1 = input("Please type in a palindrome")

    if palindrome(string1):
        print(f"{string1} is a palindrome")
        break
    else:
        print("That was not a palindrome")
```

```python

def sum_of_positives(my_list):
    positive_numbers = []
    for i in my_list:
        if i > 0:
    #add the numbers to the positive numbers list
            positive_numbers.append(i)
            
    return sum(positive_numbers)
my_list = [1, -2, 3, -4, 5]   
result = sum_of_positives(my_list)
print("The result is", result)
```

```python
def even_numbers(my_list):
    even_numbers = []
    for i in my_list:
        if i %2 == 0:
    #add the numbers to the even numbers list
            even_numbers.append(i)
            
    return even_numbers
my_list = [1, 2, 3, 4, 5]
new_list = even_numbers(my_list)
print("original", my_list)
print("new", new_list)
```

```python
def list_sum(list1,list2):
    combined_list = []

    for i in range(len(list2)):
        combined_list.append(list1[i] + list2[i])
    return combined_list


a = [1, 2, 3]
b = [7, 8, 9]

print(list_sum(a, b)) # [8, 10, 12]   
```

```python
def distinct_numbers(list1):
    new_list = []
    for number in list1:
        if number not in new_list:
            new_list.append(number)
    new_list.sort()

    return new_list

my_list = [3, 2, 2, 1, 3, 3, 1]
print(distinct_numbers(my_list)) # [1, 2, 3]
```
