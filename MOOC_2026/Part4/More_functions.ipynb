```python
def line(input_integer,input_string):
    if input_string == "":
        print( int(input_integer) * "*")
    else:
        print( int(input_integer) * input_string[0])
            
    #return ( int(input_integer) * input_string)
(line(7,"%"))
(line(3, ""))
(line(10,"LOL"))
```

```python
def greet(name):
    print("Hello there,",name)
def greet_many_times(name,times):
    while times > 0:
        greet(name)
        times -=1
greet_many_times("Emily",3)
```

```python

def box_of_hashes(height):
    for i in range (height):
        line(10,"#")
        
    

box_of_hashes(2)
print()
box_of_hashes(10)

    
```

```python

def square_of_hashes(length):
    for i in range (length):
        line(length,"#")
        
    

square_of_hashes(5)
print()
square_of_hashes(3)
```

```python
def square(length,char):
    for i in range(length):
        line(length, char)

square(5, "*")   
print()
square(3, "o")
```

```python

def triangle(rows):
    for row in range(rows):
        line(row +1,"#")
         
triangle(3)
print()
triangle(6)
```

```python
def shape(width,char,height,filler):
    for row in range(width):
        line(row +1,char)
    for row in range(height):
        line(width,filler)

shape(5,"X",3,"*")
print()
shape(2, "o", 4, "+")
print()
shape(3, ".", 0, ",")
```

```python
total_rows = 5
row = 1
while row <= total_rows:
    #print space
    space = 1
    while space <=total_rows-row:
        print(" ",end = "")
        space +=1
    #print stars
    col = 1
    while col <=(2*row-1):
        print("*",end = "")
        col +=1
    print()
    row += 1
print(" "*(total_rows - 1) + "*")
```

```python
def spruce(total_rows):
    row = 1
    print("a spruce!")
    while row <= total_rows:
        #print spaces
        space = 1
        while space<=total_rows-row:
            print(" ",end = "")
            space += 1
    #print stars
        col = 1
        while col <=(2*row-1):
            print("*",end="")
            col += 1
        print()
        row +=1
    print(" "*(total_rows - 1) + "*")
    
spruce(3)  
print()
spruce(5)
```
