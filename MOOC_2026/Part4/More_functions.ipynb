```python
def shape(width,char,height,filler):
    for row in range(width):
        line(row +1,char)
    for row in range(height):
        line(width,filler)

shape(5,"X",3,"*")
print()
shape(2, "o", 4, "+")
print()
shape(3, ".", 0, ",")
```

```python
total_rows = 5
row = 1
while row <= total_rows:
    #print space
    space = 1
    while space <=total_rows-row:
        print(" ",end = "")
        space +=1
    #print stars
    col = 1
    while col <=(2*row-1):
        print("*",end = "")
        col +=1
    print()
    row += 1
print(" "*(total_rows - 1) + "*")
```

```python
def spruce(total_rows):
    row = 1
    print("a spruce!")
    while row <= total_rows:
        #print spaces
        space = 1
        while space<=total_rows-row:
            print(" ",end = "")
            space += 1
    #print stars
        col = 1
        while col <=(2*row-1):
            print("*",end="")
            col += 1
        print()
        row +=1
    print(" "*(total_rows - 1) + "*")
    
spruce(3)  
print()
spruce(5)
```

```python
def greatest_number(a,b,c):
    """greatest_number is a function which takes in three arguments and returns the greatest in value of the three.
    """
    if a > b:
        if a > c:
            return a
    elif b > c:
        return b
    else:
        return c

print(greatest_number(3, 4, 1)) # 4
print(greatest_number(-99, -4, 7)) # 99
print(greatest_number(0, 0, 0)) # 0
        
    
```

```python
def same_chars(input_string,i,j):
    if i <0 or j < 0 or i >= len(input_string) or j >= len(input_string):
        return False
    return input_string[i] == input_string[j]
print(same_chars("programmer", 6, 7))

# different characters p and r
print(same_chars("programmer", 0, 4))

# the second index is not within the string
print(same_chars("programmer", 0, 12)) # False
```

```python
def first_word(sentence):
    return sentence.split()[0]
def second_word(sentence):
    return sentence.split()[0]


def last_word(sentence):
    return sentence.split()[-1]

sentence = "it was a dark and stormy python"
print(first_word(sentence))
print(second_word(sentence))
print(last_word(sentence))
```
