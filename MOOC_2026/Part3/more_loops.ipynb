```python
sum = 0

while True:
    number = int(input("Please type in a number , -1 to exit: "))
    if number == -1:
        break
    sum += number
    if sum > 100:
        break

print(f"The sum is {sum}")
```

```python
sum = 0
number = 0

while number != -1:
    number = int(input("Please type in a number , -1 to exit: "))
    if number != -1:
        sum += number
print(f"The sum is {sum}")
```

```python
sum = 0

while True:
    number = int(input("Please type in a number, -1 to exit: "))
    if number == -1:
        break
    if number >= 10:
        continue
    sum += number
print(f"The sum is {sum}")
```

```python
while True:
    number = int(input("Please type in a number: "))
    if number == -1:
        break
    while number > 0:
        print(number)
        number -= 1
```

```python
while True:
    number = int(input("Please type in a number: "))
    if number == -1:
        break
    while True:
        if number <= 0:
            break
        print(number)
        number -= 1
```

```python
number = int(input("Please type in a number: "))
while number > 0:
    i = 0
    while i < number:
        print(f"{i} ",end="")
        i += 1
    print()
    number -= 1
```

```python
number = int(input("Please type in a number: "))
#first create outer loop which should iterate to the number of times
iterator = 1 #iterator for outer loop
while iterator <= number:
    #tackling inner loop
    i = 1
    while i <= number:
        print(f"{iterator} x {i} = {iterator*i}")
        i +=1

    iterator += 1

```
