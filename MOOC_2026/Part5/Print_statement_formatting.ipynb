```python
name = "Mark"
age = 37
print("Hi " + name + " your age is " + str(age) + " years")
print("Hi", name, "your age is",age,"years",sep=" ")
print("Hi", name, "your age is",age,"years",sep="\n")

#setting end to an empty string means that there is no newline character at the end of the printout

print("Hi ",end= "")
print("there")

number = 1/3
print(f"The number is {number}")
print(f"The number is {number:.2f}")
```
