```python

def times_ten(start_index,end_index):
    result = {}  #initialize empty dicitonary
    for key in range(start_index, end_index + 1):
        result[key] = key * 10  #dict[key] = value
    return result
d = times_ten(3,6)
print(d)
```

```python
def factorials(n):
    result = {}

    for number in range(1, n + 1):

        factorial = 1

        for i in range(1, number + 1):
            factorial *= i

        result[number] = factorial

    return result

k = factorials(5)
print(k[1])
print(k[3])
print(k[5])
```

```python
word_list = [
  "banana", "milk", "beer", "cheese", "sourmilk", "juice", "sausage",
  "tomato", "cucumber", "butter", "margarine", "cheese", "sausage",
  "beer", "sourmilk", "sourmilk", "butter", "beer", "chocolate"
]

#how many times each word appears in the list.
def counts(my_list):
    words = {}
    for word in my_list:
        #if the word is not yet in the dcitionary,initialize the value to zero
        if word not in words:
            words[word] = 0
        #incremetn the value
        words[word] += 1
    return words
print(counts(word_list))
        
```
