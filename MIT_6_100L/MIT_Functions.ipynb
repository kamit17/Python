```python
def div_by(n,d):
    """ n and d are ints > 0
        returns True if d divides n evenly and false if odd
    """

    if n % d == 0:
        return True
    else:
        return False

print(div_by(10,3))
print(div_by(195,13))


    
```

```python
def sum_odd(a,b):
    sum_of_odds = 0
    for i in range(a,b+1):
        if i % 2 == 1:
            sum_of_odds += i

    return sum_of_odds
print(sum_odd(2,7))
```
