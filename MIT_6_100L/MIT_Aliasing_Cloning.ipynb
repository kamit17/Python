```python
#mutate L1 to remove any elements taht are also in L2.
def remove_dups(L1,L2):
    """
    remove_dups is a function that returns a list where eery element also in a second list is removed
    """
    L1_copy = L1[:]
    for e in L1_copy:
        if e in L2:
            L1.remove(e)
    return L1

list1 = [10,20,30,40]
list2 = [10,20,50,60]
remove_dups(list1,list2)
```
