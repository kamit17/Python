```python
def make_ordered_list(n):
    """ n is a posiive int
    Returns a list containing all ints in order
    from 0 to n (inclusive)
    """
    L = []
    for i in range(0,n+1):
        L.append(i)
    return L
print(make_ordered_list(6))

    
```

```python
def remove_elem(L,e):
    """
    L is a list, e is an object
    Returns a  new list with elements in the same order as L
    but without any elements equal to e.
    """
    newlist = []
    for i in L:
        # i 1 , then 2, then 2 then 2
        if i != e:
            newlist.append(i)
    return newlist


L = [1,2,2,2]
#print(remove_elem(L,2))   #prints 1
#print(remove_elem(L,1))
print(remove_elem(L,0))
```

```python
def count_words(sen):
    """sen is a string representing a sentence
    Returns how many words are in a sentence(i.e. a word is a 
    sequence of characters between space."""

    L = sen.split(' ')
    return len(L)

s = "Hello it is me"
print(count_words(s))
s1 = "sen is a string representing a sentence .Returns how many words are in a sentence(i.e. a word is a sequence of characters between space"
        
print(count_words(s1))    
```

```python
def square_list(L):
    for i in range(len(L)):
        L[i] = L[i] ** 2
    return L
    #print(L)
List1 = [2,3,4]
print(f"before function call:,{List1}")
square_list(List1)
print(f"After function call:,{List1}")
    
```

```python
#range returns something that behaves like a tuple (but isn't - it retruns an iterable)
# A loop iterates over indices of L and mutates L each time.
#range(4)  __> Kind of like tuple(0,1,2,3)
L = [1,2,3,4]
for i in range(len(L)):
    L.append(i)
    print(f"{[i]} th time")
    print(L)
```

```python
#This results in a indefinite loop as it iterates over L directly

L = [1,2,3,4]
i = 0
for e in L:
    L.append(i)
    i += 1
    print(L)
```
