```python
#Decimal to binary conversion

num = int(input("Enter any number: "))

if num < 0:
    is_neg = True
    num = abs(num)
else:
    is_neg = False

result = ' '
if num == 0:
    result = '0'
while num > 0:
    result = str(num%2) + result
    num = num //2
if is_neg:
    result = "-" + result
print(f"Binary representation is {result}")
```

```python
#x = 0.625
x = float(input('Enter a decimal number between 0 and 1: '))
p = 0

#Find power of 2 to make an integer
while((2**p)*x)%1 != 0:      #% grabs the decimal part only
    print('Remainder = ' +str((2**p)*x - int((2**p)*x)))
    p += 1
    
###Convert to int
num = int(x*(2**p))

#Encode as binary number
result = ''
if num == 0:
    result = '0'
while num > 0:
    result = str(num%2) + result
    num = num//2

#pad front with 0's i.e shift right
for i in range(p - len(result)):
    result = '0' + result

#insert decimal
result = result[0:-p] + '.' + result[-p:]

print(f"The binary representation of {str(x)} is {str(result)}")
```

```python
#Approximation

x = 54321
epsilon = 0.01 # how close we want to be to the answer
num_guesses = 0   #keep track of how many guesses we are doing
guess = 0.0
increment = 0.00001  #step incremeent of guess

while abs(guess**2 - x) >= epsilon and guess**2 <= x:
    guess += increment
    num_guesses += 1
#    if num_guesses%100000 == 0:
#        print(f"Current guess = {guess}")
#        print(f"Current guess**2 -x  = {abs(guess*guess -x)}")

print(f"num_guesses = {num_guesses}")
#print(f"{guess} is close to square root of {x}") 

if abs(guess**2 -x ) >= epsilon:
    print(f"Failed on square root of {x}")
else:
    print(f"{guess} is close to square root of {x}")
    
```
